## Notebook to run the analysis for the long-term adiabatic expansions used in Section 5 and 5.1 of the paper Fig.7 and Fig. 8


```python
from expansion_tools.blob_expansion_tools_jetset import *
from expansion_tools.delay_analysis import gamma_radio_delay_fit,gamma_radio_delay_analysis_vs_v_exp,gamma_radio_delay_analysis_vs_freq
from jetset.data_loader import Data,ObsData
from jetset.plot_sedfit import PlotSED
```

```python
from jetset.jet_timedep import JetTimeEvol
temp_ev_acc_flare=JetTimeEvol.load_model('temp_ev_sim_flare/temp_ev_acc_flare.pkl')
R0=temp_ev_acc_flare.rad_region.jet.parameters.R.val
delta=temp_ev_acc_flare.rad_region.jet.parameters.beam_obj.val

temp_ev_exp=JetTimeEvol.load_model('temp_ev_sim_exp_change_beta/temp_ev_expansion_beta_exp=0.001.pkl')
t_exp=temp_ev_exp.parameters.t_jet_exp.val

t_exp_obs_d=t_exp/86400/delta
print(t_exp_obs_d,R0)
```

```python
from expansion_tools.delay_analysis import gamma_radio_delay_analysis_vs_v_exp
from astropy.table import Table
from astropy.units import Unit as u
import glob
file_list=glob.glob('temp_ev_sim_exp_change_beta/lc_obs_beta_exp=*.pkl')
file_list.sort()
beta_exp=np.zeros(len(file_list))
delta_t=np.zeros(len(file_list))
delta_t_conv=np.zeros(len(file_list))
delta_t_conv_err=np.zeros(len(file_list))
t_decay=np.zeros(len(file_list))
t_decay_conv=np.zeros(len(file_list))
t_decay_conv_err=np.zeros(len(file_list))
t_rise_conv=np.zeros(len(file_list))
t_rise_conv_err=np.zeros(len(file_list))
fig_list=[None]*len(file_list)


for ID,f in enumerate(file_list):
    beta_exp[ID],delta_t[ID],delta_t_conv[ID],delta_t_conv_err[ID],t_decay[ID],t_decay_conv[ID],t_decay_conv_err[ID],t_rise_conv[ID],t_rise_conv_err[ID],fig_list[ID]= gamma_radio_delay_fit(f,'lg_obs','lr_obs_15GHz',flare_lc='temp_ev_sim_flare/lc_obs_flare.pkl',t_start_frac=0.5,t_stop_frac=0.1,phi_frozen=True,psi_frozen=True,max_points=5000,delta=delta,R0=R0,t_exp=t_exp)
    print('ID=%d bexp=%f'%(ID,beta_exp[ID]))
    print('='*80)
    print('')
    print('')
t = Table([beta_exp*u('c-1'),delta_t*u('s'),delta_t_conv*u('s'),delta_t_conv_err*u('s'),t_decay*u('s'),t_decay_conv*u('s'),t_decay_conv_err*u('s'),t_rise_conv*u('s'),t_rise_conv_err*u('s')], names=('beta_exp', 'delta_T', 'delta_T_conv','delta_T_conv_err','t_decay','t_decay_conv','t_decay_conv_err','t_rise_conv','t_rise_conv_err'))
t.write('temp_ev_analysis_beta_exp_trends/gamma_radio_delay_fit_vs_beta_exp.ecsv',format='ascii.ecsv')
```

```python
ID=0
fig_list[ID].axes[0].set_title('')
fig_list[ID].tight_layout()
fig_list[ID].savefig('temp_ev_analysis_beta_exp_trends/conv_example_bexp_%1.3f_15GHz.pdf'%beta_exp[ID])
fig_list[ID]
```
