```python
lc_file='temp_ev_sim_exp_change_beta/lc_obs_beta_exp=0.001.pkl'
import pickle
with open(lc_file, 'rb') as f:
    l=pickle.load(f)
l.keys()
```

```python
beta_exp_sim=np.float(lc_file.split('=')[1].replace('.pkl',''))
print('beta_exp_sim',beta_exp_sim)
```

```python
radio_keys=[k for k in l.keys() if 'lr_obs' in k]
radio_keys
```

```python
beta_exp=np.zeros(len(radio_keys))
delta_t=np.zeros(len(radio_keys))
delta_t_conv=np.zeros(len(radio_keys))
delta_t_conv_err=np.zeros(len(radio_keys))
t_decay=np.zeros(len(radio_keys))
t_decay_conv=np.zeros(len(radio_keys))
t_decay_conv_err=np.zeros(len(radio_keys))
t_rise_conv=np.zeros(len(radio_keys))
t_rise_conv_err=np.zeros(len(radio_keys))
nu=np.zeros(len(radio_keys))
fig_list=[None]*len(radio_keys)
                     
fig_list=[None]*len(radio_keys)
for ID,k in enumerate(radio_keys):
    nu[ID]=1E9*np.float(k.split('_')[2].replace('GHz',''))
    if nu[ID]<45E9:
        beta_exp[ID],delta_t[ID],delta_t_conv[ID],delta_t_conv_err[ID],t_decay[ID],t_decay_conv[ID],t_decay_conv_err[ID],t_rise_conv[ID],t_rise_conv_err[ID],fig_list[ID]= gamma_radio_delay_fit(lc_file,'lg_obs',k,flare_lc='temp_ev_sim_flare/lc_obs_flare.pkl',t_start_frac=0.55,t_stop_frac=0.1,phi_frozen=True,psi_frozen=True,max_points=5000,delta=delta,R0=R0,t_exp=t_exp,beta_exp=beta_exp_sim,flare_duration=None)
    
    
t = Table([nu*u('Hz'),beta_exp*u('c-1'),delta_t*u('s'),delta_t_conv*u('s'),delta_t_conv_err*u('s'),t_decay*u('s'),t_decay_conv*u('s'),t_decay_conv_err*u('s'),t_rise_conv*u('s'),t_rise_conv_err*u('s')], names=('nu_obs','beta_exp', 'delta_T', 'delta_T_conv','delta_T_conv_err','t_decay','t_decay_conv','t_decay_conv_err','t_rise_conv','t_rise_conv_err'))
t.write('temp_ev_analysis_nu_trends/gamma_radio_delay_fit_vs_freq.ecsv',format='ascii.ecsv')
```
